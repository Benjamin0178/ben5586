```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url = 'https://www.marketwatch.com/investing/stock/tsla/financials'
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup (page.text, 'html')
```

```python
soup.find_all('div', class_ = 'cell__content fixed--cell')
```

```python
revenue = soup.find_all('div', class_ = 'cell__content fixed--cell')
```

```python
for i in revenue:
        print(i.text)
```

```python
price =soup.find_all ('div', class_ = 'cell__content')
```

```python
for i in price:
    print(i.text)
```
